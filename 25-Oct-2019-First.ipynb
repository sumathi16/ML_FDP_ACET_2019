```python
%matplotlib inline
# Data Analysis
import pandas as pd
import numpy as np
#Plotting
import matplotlib.pyplot as plt
import seaborn as sns
# Machine Learning Models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
# Metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

Read the data Display first 5 rows

```python
df = pd.read_csv('Datasets/titanic.csv')
df.head()
```

Check the missing values

```python
df.isnull().sum()
```

There are three columns which have missing values<br>
Remove the Cabin Column<br>
Replace the missing values in Age with mean<br>
Remove the two rows of the remaining missing values

```python
df.drop('Cabin',axis=1,inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df.shape
```

Check whether the missing values in the Age columns are replaced with mean or not

```python
print(df['Age'][5])
print(df['Age'].isnull().sum())
```

Now remove the rows with missing values

```python
df.dropna(inplace=True)
print(df.shape)
print(df.isnull().sum().sum())
```

Assign the names columns to a new variable<br>
Remove the columns Name,Passenger Id, Ticket

```python
df.columns
```

```python
names = df['Name']
df.drop(['Name','Ticket','PassengerId'],axis=1,inplace=True)
df.shape
```

Check the categorical Columns

```python
(df.dtypes == object).sum()
```

Caluculate the unique values in the categorical columns

```python
print(df['Sex'].nunique())
print(df['Embarked'].nunique())
```

Convert the columns with two unique values using Label Binarizer

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
lb.fit_transform(df['Sex'])
```

```python
#lb.fit_transform(df['Sex']).reshape(-1)
df['Sex'] = lb.fit_transform(df['Sex']).flatten()
df['Sex'].dtype
```

Apply Label Encoder for the categorical columns 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Embarked'] = le.fit_transform(df['Embarked'])
df['Embarked'][:3] 
```

```python
le.classes_
```

```python
df.describe()
```

Separate the output column

```python
target = df.Survived
features = df.drop('Survived',axis=1)
```

Apply Minmax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
features = pd.DataFrame(sc.fit_transform(features),
                        columns= features.columns)
features.describe()
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(features,target,
                                             random_state=42)
xtrain.head()
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
ytest_pred  = lr.predict(xtest) 
ytrain_pred = lr.predict(xtrain)
print('----------Accuracy-------')
print('training',accuracy_score(ytrain_pred,ytrain))
print('testing',accuracy_score(ytest_pred,ytest))
print('----------f1score-------')
print('training',f1_score(ytrain_pred,ytrain))
print('testing',f1_score(ytest_pred,ytest))
```

```python
knn = KNeighborsClassifier()
knn.fit(xtrain,ytrain)
ytest_pred  = knn.predict(xtest) 
ytrain_pred = knn.predict(xtrain)
print('----------Accuracy-------')
print('training',accuracy_score(ytrain_pred,ytrain))
print('testing',accuracy_score(ytest_pred,ytest))
print('----------f1score-------')
print('training',f1_score(ytrain_pred,ytrain))
print('testing',f1_score(ytest_pred,ytest))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
scoring = make_scorer(accuracy_score)
params = {"n_neighbors":[3,7,9,11,13],
          "weights":['uniform','distance']}
grid = GridSearchCV(KNeighborsClassifier(),params,scoring)
grid.fit(xtrain,ytrain)
```

```python
grid.best_estimator_
```

```python
accuracy_score(grid.predict(xtest),ytest)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(grid,features,target,cv=5)
```
