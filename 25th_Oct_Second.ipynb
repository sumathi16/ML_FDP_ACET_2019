```python
plt.figure(figsize=(16,40))
num = df.columns[df.dtypes==np.float64]
for i in range(len(num)):
    plt.subplot(6,2,i+1)
    plt.boxplot(df[num[i]])
```

```python
new_df = df[num]
new_df.shape
```

```python
new_df.head()
```

```python
from sklearn.cluster import KMeans
mod = KMeans(2)
mod.fit(new_df)
```

```python
pred = mod.predict(new_df)
```

```python
df['color']
```

```python
df['pred'] = pred
```

```python
df.groupby(['color','pred'])['pred'].count()
```
